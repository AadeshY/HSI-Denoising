```python
# ============================
# Hyperspectral Image Denoising
# ============================

# ----------------------------
# 1. Install Necessary Libraries
# ----------------------------
# Uncomment and run the following lines if you haven't installed these packages.
# !pip uninstall bm3d -y
# !pip uninstall bm4d -y
# !pip install bm3d
# !pip install torch torchvision
# !pip install scikit-image
# !pip install stable-baselines3
# !pip install gymnasium
# !pip install seaborn

# ----------------------------
# 2. Import Necessary Libraries
# ----------------------------
import os
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr, structural_similarity as ssim
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import PPO
import seaborn as sns
import pandas as pd
from torchvision import transforms
import logging

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Verify scikit-image installation
import skimage
print(f"scikit-image version: {skimage.__version__}")

# ----------------------------
# 3. Define Utility Functions
# ----------------------------
def compute_sam(clean, denoised):
    """
    Compute the Spectral Angle Mapper (SAM) between two images.

    Parameters:
        clean (numpy.ndarray): Clean image with shape (H, W, C).
        denoised (numpy.ndarray): Denoised image with shape (H, W, C).

    Returns:
        sam (float): Average SAM over all pixels in degrees.
    """
    h, w, c = clean.shape
    clean_flat = clean.reshape(-1, c)
    denoised_flat = denoised.reshape(-1, c)
    dot_product = np.sum(clean_flat * denoised_flat, axis=1)
    norm_clean = np.linalg.norm(clean_flat, axis=1)
    norm_denoised = np.linalg.norm(denoised_flat, axis=1)
    cos_sim = dot_product / (norm_clean * norm_denoised + 1e-8)
    cos_sim = np.clip(cos_sim, -1.0, 1.0)
    angles = np.arccos(cos_sim) * (180.0 / np.pi)
    sam = np.mean(angles)
    return sam

def visualize_results(noisy, clean, denoised_tfpnp, denoised_qrnn3d, sample=0, bands=[0,1,2]):
    """
    Visualize noisy, denoised, and clean images for specified bands of a single sample.

    Parameters:
        noisy (numpy.ndarray): Noisy image batches with shape [B, C, H, W].
        clean (numpy.ndarray): Clean image batches with shape [B, C, H, W].
        denoised_tfpnp (numpy.ndarray): Denoised image batches from TFPNP with shape [B, C, H, W].
        denoised_qrnn3d (numpy.ndarray): Denoised image batches from QRNN3D with shape [B, C, H, W].
        sample (int): Index of the sample in the batch to visualize.
        bands (list): List of band indices to visualize.
    """
    num_bands = len(bands)
    plt.figure(figsize=(15, 5))

    for i, band in enumerate(bands):
        plt.subplot(1, num_bands * 3, i * 3 + 1)
        plt.imshow(noisy[sample, band, :, :], cmap='gray')
        plt.title(f'Noisy - Band {band}')
        plt.axis('off')

        plt.subplot(1, num_bands * 3, i * 3 + 2)
        plt.imshow(denoised_tfpnp[sample, band, :, :], cmap='gray')
        plt.title('TFPNP Denoised')
        plt.axis('off')

        plt.subplot(1, num_bands * 3, i * 3 + 3)
        plt.imshow(denoised_qrnn3d[sample, band, :, :], cmap='gray')
        plt.title('QRNN3D Denoised')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

def visualize_batch_samples(noisy, clean, denoised, bands=[0,1,2], num_samples=3):
    """
    Visualize noisy, denoised, and clean images for multiple samples and specified bands.

    Parameters:
        noisy (numpy.ndarray): Noisy image batches with shape [B, C, H, W].
        clean (numpy.ndarray): Clean image batches with shape [B, C, H, W].
        denoised (numpy.ndarray): Denoised image batches with shape [B, C, H, W].
        bands (list): List of band indices to visualize.
        num_samples (int): Number of samples in the batch to visualize.
    """
    num_bands = len(bands)
    plt.figure(figsize=(15, 5 * num_samples))

    for sample_idx in range(num_samples):
        for i, band in enumerate(bands):
            plt.subplot(num_samples, num_bands * 3, sample_idx * num_bands * 3 + i * 3 + 1)
            plt.imshow(noisy[sample_idx, band, :, :], cmap='gray')
            plt.title(f"Noisy - Sample {sample_idx + 1} - Band {band}")
            plt.axis('off')

            plt.subplot(num_samples, num_bands * 3, sample_idx * num_bands * 3 + i * 3 + 2)
            plt.imshow(denoised[sample_idx, band, :, :], cmap='gray')
            plt.title(f"Denoised - Sample {sample_idx + 1} - Band {band}")
            plt.axis('off')

            plt.subplot(num_samples, num_bands * 3, sample_idx * num_bands * 3 + i * 3 + 3)
            plt.imshow(clean[sample_idx, band, :, :], cmap='gray')
            plt.title(f"Clean - Sample {sample_idx + 1} - Band {band}")
            plt.axis('off')

    plt.tight_layout()
    plt.show()

def save_visualization(noisy, clean, denoised, bands=[0,1,2], sample_idx=0, filename='denoising_result.png'):
    """
    Save the visualization of noisy, denoised, and clean images for specified bands of a single sample.

    Parameters:
        noisy (numpy.ndarray): Noisy image batches with shape [B, C, H, W].
        clean (numpy.ndarray): Clean image batches with shape [B, C, H, W].
        denoised (numpy.ndarray): Denoised image batches with shape [B, C, H, W].
        bands (list): List of band indices to visualize.
        sample_idx (int): Index of the sample in the batch to visualize.
        filename (str): Filename to save the visualization.
    """
    num_bands = len(bands)
    plt.figure(figsize=(15, 5))

    for i, band in enumerate(bands):
        plt.subplot(1, num_bands * 3, i * 3 + 1)
        plt.imshow(noisy[sample_idx, band, :, :], cmap='gray')
        plt.title(f"Noisy - Band {band}")
        plt.axis('off')

        plt.subplot(1, num_bands * 3, i * 3 + 2)
        plt.imshow(denoised[sample_idx, band, :, :], cmap='gray')
        plt.title(f"Denoised - Band {band}")
        plt.axis('off')

        plt.subplot(1, num_bands * 3, i * 3 + 3)
        plt.imshow(clean[sample_idx, band, :, :], cmap='gray')
        plt.title(f"Clean - Band {band}")
        plt.axis('off')

    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    plt.close()

# ----------------------------
# 4. Define Dataset Class
# ----------------------------
class HSIDataset(Dataset):
    def __init__(self, noisy_patches, clean_patches, augment=False):
        """
        Hyperspectral Image Dataset.

        Parameters:
            noisy_patches (numpy.ndarray): Noisy image patches.
            clean_patches (numpy.ndarray): Corresponding clean image patches.
            augment (bool): Whether to apply data augmentation.
        """
        self.noisy_patches = noisy_patches
        self.clean_patches = clean_patches
        self.augment = augment
        if self.augment:
            self.transform = transforms.Compose([
                transforms.ToPILImage(),
                transforms.RandomHorizontalFlip(),
                transforms.RandomVerticalFlip(),
                transforms.RandomRotation(90),
                transforms.ToTensor(),
            ])
        else:
            self.transform = None

    def __len__(self):
        return len(self.noisy_patches)

    def __getitem__(self, idx):
        noisy = self.noisy_patches[idx].transpose(2, 0, 1)  # Shape: [C, H, W]
        clean = self.clean_patches[idx].transpose(2, 0, 1)
        if self.augment:
            # Apply the same transformation to both noisy and clean patches
            seed = np.random.randint(2147483647)  # Get a random seed
            torch.manual_seed(seed)
            noisy = self.transform(torch.from_numpy(noisy)).numpy()
            torch.manual_seed(seed)
            clean = self.transform(torch.from_numpy(clean)).numpy()
        else:
            noisy = torch.from_numpy(noisy).float()
            clean = torch.from_numpy(clean).float()
        return noisy, clean

# ----------------------------
# 5. Data Preparation Function
# ----------------------------
def download_and_preprocess_data(noise_type='gaussian', noise_params={}):
    """
    Download, normalize, and preprocess the Indian Pines dataset.

    Parameters:
        noise_type (str): Type of noise to add ('gaussian', 'salt_pepper').
        noise_params (dict): Parameters for the noise type.

    Returns:
        patches_noisy (numpy.ndarray): Noisy image patches.
        patches_clean (numpy.ndarray): Clean image patches.
    """
    # Download the Indian Pines dataset if not already present
    dataset_url = "http://www.ehu.eus/ccwintco/uploads/6/67/Indian_pines_corrected.mat"
    dataset_path = "indian_pines_corrected.mat"
    if not os.path.exists(dataset_path):
        logging.info(f"Downloading dataset from {dataset_url}...")
        os.system(f"wget '{dataset_url}' -O {dataset_path}")
    else:
        logging.info("Dataset already downloaded.")

    # Load and normalize the dataset
    logging.info("Loading Indian Pines dataset...")
    data = scipy.io.loadmat(dataset_path)
    hsi = data['indian_pines_corrected'].astype('float32')
    hsi_normalized = (hsi - hsi.min()) / (hsi.max() - hsi.min())
    logging.info(f"Original Dataset Shape: {hsi_normalized.shape}")

    # Select specific channels (29, 60, 90)
    selected_channels = [29, 60, 90]
    hsi_selected = hsi_normalized[:, :, selected_channels]
    logging.info(f"Dataset Shape After Channel Selection: {hsi_selected.shape}")

    # Add specified noise type
    logging.info(f"Adding {noise_type} noise to dataset...")
    noisy_hsi = add_noise(hsi_selected, noise_type=noise_type, **noise_params)
    logging.info("Noise added and data clipped to [0,1].")

    # Extract patches
    logging.info("Extracting patches...")
    patches_noisy = extract_patches(noisy_hsi, patch_size=64, stride=16)  # Increased patch_size to 64
    patches_clean = extract_patches(hsi_selected, patch_size=64, stride=16)  # Increased patch_size to 64
    logging.info(f"Total patches extracted: {len(patches_noisy)}")
    return patches_noisy, patches_clean

def extract_patches(data, patch_size=64, stride=16):
    """
    Extract overlapping patches from the data.

    Parameters:
        data (numpy.ndarray): Input data.
        patch_size (int): Size of each patch.
        stride (int): Stride for patch extraction.

    Returns:
        patches (numpy.ndarray): Extracted patches.
    """
    patches = []
    h, w = data.shape[:2]
    for i in range(0, h - patch_size + 1, stride):
        for j in range(0, w - patch_size + 1, stride):
            patch = data[i:i + patch_size, j:j + patch_size]
            if patch.shape[0] == patch_size and patch.shape[1] == patch_size:
                patches.append(patch)
    patches = np.array(patches)

    # Assert that all patches are of correct size
    assert patches.shape[1] >= 8 and patches.shape[2] >= 8, "One or more patches are smaller than BM3D block size."

    return patches

def add_noise(data, noise_type='gaussian', **kwargs):
    """
    Add different types of noise to the data.

    Parameters:
        data (numpy.ndarray): Original clean data.
        noise_type (str): Type of noise ('gaussian', 'salt_pepper', etc.).
        **kwargs: Additional parameters for noise types.

    Returns:
        noisy_data (numpy.ndarray): Data with added noise.
    """
    if noise_type == 'gaussian':
        noise_std = kwargs.get('noise_std', 0.1)
        noisy_data = data + noise_std * np.random.randn(*data.shape)
        noisy_data = np.clip(noisy_data, 0, 1)
    elif noise_type == 'salt_pepper':
        amount = kwargs.get('amount', 0.05)
        noisy_data = data.copy()
        num_salt = np.ceil(amount * data.size * 0.5).astype(int)
        num_pepper = np.ceil(amount * data.size * 0.5).astype(int)
        # Salt
        coords = [np.random.randint(0, i, num_salt) for i in data.shape]
        noisy_data[tuple(coords)] = 1
        # Pepper
        coords = [np.random.randint(0, i, num_pepper) for i in data.shape]
        noisy_data[tuple(coords)] = 0
    else:
        raise ValueError("Unsupported noise type.")
    return noisy_data

# ----------------------------
# 6. Define Denoiser Models
# ----------------------------

# a. Residual Block
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        """
        Residual Block with two convolutional layers.

        Parameters:
            in_channels (int): Number of input channels.
            out_channels (int): Number of output channels.
        """
        super(ResidualBlock, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.bn1   = nn.BatchNorm2d(out_channels)
        self.relu  = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1)
        self.bn2   = nn.BatchNorm2d(out_channels)

        if in_channels != out_channels:
            self.residual = nn.Conv2d(in_channels, out_channels, kernel_size=1)
        else:
            self.residual = nn.Identity()

    def forward(self, x):
        identity = self.residual(x)
        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += identity
        out = self.relu(out)
        return out

# b. UNet Components
class DoubleConv(nn.Module):
    """Applies two consecutive convolutional layers each followed by ReLU activation."""
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),  # Padding to maintain spatial dimensions
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

class Down(nn.Module):
    """Downscaling with maxpool followed by double convolution."""
    def __init__(self, in_channels, out_channels):
        super(Down, self).__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),  # Reduces spatial dimensions by a factor of 2
            DoubleConv(in_channels, out_channels)
        )

    def forward(self, x):
        return self.maxpool_conv(x)

class Up(nn.Module):
    """Upscaling followed by double convolution."""
    def __init__(self, in_channels, out_channels, bilinear=True):
        super(Up, self).__init__()

        # If bilinear, use the normal convolutions to reduce the number of channels after upsampling
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels)
        else:
            # Learnable up-convolution
            self.up = nn.ConvTranspose2d(in_channels // 2, in_channels // 2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)

        # Input is CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        # Padding to handle any mismatch in dimensions
        x1 = nn.functional.pad(x1, [diffX // 2, diffX - diffX // 2,
                                    diffY // 2, diffY - diffY // 2])

        # Concatenate along the channels axis
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)

class OutConv(nn.Module):
    """Final convolution to map to the desired number of output channels."""
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        return self.conv(x)

# c. Residual UNet Denoiser
class ResidualUNetDenoiser(nn.Module):
    def __init__(self, n_channels=3, n_classes=3, bilinear=True):
        """
        Residual UNet Denoiser with deeper architecture and skip connections.

        Parameters:
            n_channels (int): Number of input channels.
            n_classes (int): Number of output channels.
            bilinear (bool): Whether to use bilinear upsampling.
        """
        super(ResidualUNetDenoiser, self).__init__()
        self.n_channels = n_channels
        self.n_classes  = n_classes
        self.bilinear   = bilinear

        # Encoder
        self.inc = ResidualBlock(n_channels, 64)
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        factor = 2 if bilinear else 1
        self.down4 = Down(512, 1024 // factor)

        # Decoder
        self.up1 = Up(1024, 512 // factor, bilinear)
        self.up2 = Up(512, 256 // factor, bilinear)
        self.up3 = Up(256, 128 // factor, bilinear)
        self.up4 = Up(128, 64, bilinear)
        self.outc = OutConv(64, n_classes)
        self.sigmoid = nn.Sigmoid()  # Clamp outputs to [0,1]

    def forward(self, x):
        x1 = self.inc(x)          # [B, 64, H, W]
        x2 = self.down1(x1)       # [B, 128, H/2, W/2]
        x3 = self.down2(x2)       # [B, 256, H/4, W/4]
        x4 = self.down3(x3)       # [B, 512, H/8, W/8]
        x5 = self.down4(x4)       # [B, 1024, H/16, W/16]
        x = self.up1(x5, x4)      # [B, 512, H/8, W/8]
        x = self.up2(x, x3)       # [B, 256, H/4, W/4]
        x = self.up3(x, x2)       # [B, 128, H/2, W/2]
        x = self.up4(x, x1)       # [B, 64, H, W]
        logits = self.outc(x)     # [B, 3, H, W]
        return self.sigmoid(logits)

# d. QRNN3D Denoiser (Simplified Placeholder)
class QRNNREDC3D(nn.Module):
    def __init__(self, in_channels, channels, num_half_layer, sample_idx, BiQRNNConv3D=None, BiQRNNDeConv3D=None,
                 QRNN3DEncoder=None, QRNN3DDecoder=None, is_2d=False, has_ad=True, bn=True, act='tanh', plain=False):
        super(QRNNREDC3D, self).__init__()
        # Simplified QRNN3D model. Replace with full implementation as needed.
        self.denoiser = nn.Sequential(
            nn.Conv3d(in_channels=in_channels, out_channels=channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv3d(in_channels=channels, out_channels=in_channels, kernel_size=3, padding=1)
        )

    def forward(self, x):
        return self.denoiser(x) + x  # Residual connection

# ----------------------------
# 7. Define ADMM Solver
# ----------------------------
class ADMMSolver(nn.Module):
    def __init__(self, denoiser, max_iterations=10):
        """
        ADMM Solver integrating the denoiser within an iterative optimization framework.

        Parameters:
            denoiser (nn.Module): Denoising neural network.
            max_iterations (int): Maximum number of ADMM iterations.
        """
        super(ADMMSolver, self).__init__()
        self.denoiser = denoiser
        self.max_iterations = max_iterations
        self.is_3d = self.check_if_3d()

    def check_if_3d(self):
        """
        Check if the denoiser contains any 3D convolutional layers.

        Returns:
            bool: True if 3D convolutions are present, False otherwise.
        """
        for layer in self.denoiser.modules():
            if isinstance(layer, nn.Conv3d):
                return True
        return False

    def forward(self, noisy, params, iterations=None):
        """
        Forward pass through the ADMM Solver.

        Parameters:
            noisy (torch.Tensor): Noisy input image tensor with shape [B, C, H, W].
            params (torch.Tensor): Parameters tensor containing sigma_d and mu with shape [B, 2].
            iterations (int or list, optional): Number of ADMM iterations. If int, applies to all samples.
                                               If list, applies per sample. Defaults to self.max_iterations.

        Returns:
            torch.Tensor: Denoised image tensor with shape [B, C, H, W].
        """
        if iterations is None:
            iterations = self.max_iterations

        x = noisy.clone()
        z = noisy.clone()
        u = torch.zeros_like(noisy)

        # params shape: [batch_size, 2]
        sigma_d = params[:, 0].view(-1, 1, 1, 1)  # [B, 1, 1, 1]
        mu = params[:, 1].view(-1, 1, 1, 1)       # [B, 1, 1, 1]

        if isinstance(iterations, list):
            max_iter = max(iterations)
        else:
            max_iter = iterations

        for iter_num in range(max_iter):
            if self.is_3d:
                # Reshape to [B, C, D=1, H, W]
                x_input = z - u
                logging.debug(f"Before reshaping for 3D: {x_input.shape}")  # Debugging
                x_input = x_input.unsqueeze(2)  # Adds a depth dimension
                logging.debug(f"After reshaping for 3D: {x_input.shape}")   # Debugging
                denoised = self.denoiser(x_input)
                logging.debug(f"Denoised shape after 3D denoiser: {denoised.shape}")  # Debugging
                denoised = denoised.squeeze(2)  # Removes the depth dimension
                logging.debug(f"Denoised shape after squeezing: {denoised.shape}")      # Debugging
            else:
                denoised = self.denoiser(z - u)
                logging.debug(f"Denoised shape for 2D denoiser: {denoised.shape}")    # Debugging

            x = denoised  # Update x
            z = (x + u) / (1 + mu)  # Update z
            u = u + (x - z)         # Update u

        return x

# ----------------------------
# 8. Define RL Environment
# ----------------------------
class ADMMEnv(gym.Env):
    """
    Custom Environment for ADMM Solver with RL-based parameter tuning.
    """
    def __init__(self, model, device, test_loader, max_iterations=10):
        super(ADMMEnv, self).__init__()
        self.model = model
        self.device = device
        self.test_loader = test_loader
        self.max_iterations = max_iterations
        self.current_step = 0
        self.total_psnr = 0
        self.data_iter = iter(self.test_loader)
        self.observation_space = spaces.Box(low=0.0, high=100.0, shape=(1,), dtype=np.float32)
        self.action_space = spaces.Box(low=np.array([0.0, 0.0, 1]),
                                       high=np.array([1.0, 1.0, self.max_iterations]),
                                       dtype=np.float32)

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.total_psnr = 0
        initial_psnr = 0.0
        observation = np.array([initial_psnr], dtype=np.float32)
        info = {}
        return observation, info

    def step(self, action):
        sigma_d, mu, tau = action
        sigma_d = np.clip(sigma_d, 0.0, 1.0)
        mu = np.clip(mu, 0.0, 1.0)
        tau = int(np.clip(tau, 1, self.max_iterations))

        params = torch.tensor([[sigma_d, mu]], dtype=torch.float32).to(self.device)

        # Get one batch from test_loader
        try:
            noisy, clean = next(self.data_iter)
        except StopIteration:
            self.data_iter = iter(self.test_loader)
            noisy, clean = next(self.data_iter)

        noisy, clean = noisy.to(self.device), clean.to(self.device)

        denoised = self.model(noisy, params=params, iterations=tau)
        denoised = denoised.detach().cpu().numpy()[0].transpose(1,2,0)  # [H, W, C]
        clean_np = clean.cpu().numpy()[0].transpose(1,2,0)             # [H, W, C]

        # Compute PSNR as reward
        reward = psnr(clean_np, denoised, data_range=1)
        self.current_psnr = reward

        done = True  # Episode ends after one step
        info = {}

        observation = np.array([self.current_psnr], dtype=np.float32)
        return observation, reward, done, False, info

    def render(self, mode='human'):
        pass  # Implement visualization if needed

    def close(self):
        pass  # Perform any necessary cleanup

# ----------------------------
# 9. Initialize Models and PPO Agents
# ----------------------------
def initialize_models_and_agents(device, test_loader):
    """
    Initialize denoiser models, ADMM solvers, environments, and PPO agents.

    Parameters:
        device (torch.device): Device to run computations on.
        test_loader (DataLoader): DataLoader for the test set.

    Returns:
        admm_tfpnp (ADMMSolver): ADMM Solver with TFPNP denoiser.
        admm_qrnn3d (ADMMSolver): ADMM Solver with QRNN3D denoiser.
        ppo_tfpnp (PPO): PPO agent for TFPNP.
        ppo_qrnn3d (PPO): PPO agent for QRNN3D.
        optimizer_tfpnp (torch.optim.Optimizer): Optimizer for TFPNP.
        optimizer_qrnn3d (torch.optim.Optimizer): Optimizer for QRNN3D.
    """
    # Initialize TFPNP's UNet
    tfpn_unet = ResidualUNetDenoiser(n_channels=3, n_classes=3, bilinear=True).to(device)
    admm_tfpnp = ADMMSolver(denoiser=tfpn_unet, max_iterations=10).to(device)
    optimizer_tfpnp = optim.Adam(admm_tfpnp.parameters(), lr=1e-4)

    # Initialize QRNN3D with corrected in_channels
    qrnn3d = QRNNREDC3D(
        in_channels=3,  # Correct number of input channels
        channels=64,
        num_half_layer=3,
        sample_idx=[1,2]
    ).to(device)
    admm_qrnn3d = ADMMSolver(denoiser=qrnn3d, max_iterations=10).to(device)
    optimizer_qrnn3d = optim.Adam(admm_qrnn3d.parameters(), lr=1e-4)

    # Initialize Environments
    env_tfpnp = ADMMEnv(admm_tfpnp, device, test_loader, max_iterations=10)
    env_qrnn3d = ADMMEnv(admm_qrnn3d, device, test_loader, max_iterations=10)

    # Initialize PPO Agents
    ppo_tfpnp = PPO('MlpPolicy', env_tfpnp, verbose=1)
    ppo_qrnn3d = PPO('MlpPolicy', env_qrnn3d, verbose=1)

    return admm_tfpnp, admm_qrnn3d, ppo_tfpnp, ppo_qrnn3d, optimizer_tfpnp, optimizer_qrnn3d

# ----------------------------
# 10. Evaluation Functions
# ----------------------------
def evaluate_model(model, denoiser, device, test_loader):
    """
    Evaluate the ADMM Solver with a denoiser on the test set.

    Parameters:
        model (ADMMSolver): The ADMM Solver model.
        denoiser (nn.Module): The denoising neural network.
        device (torch.device): Device to run computations on.
        test_loader (DataLoader): DataLoader for the test set.

    Returns:
        mean_psnr, mean_ssim, mean_sam: Evaluation metrics.
    """
    model.eval()
    psnr_list, ssim_list, sam_list = [], [], []
    with torch.no_grad():
        for noisy, clean in test_loader:
            noisy, clean = noisy.to(device), clean.to(device)
            # Example params: fixed sigma_d and mu
            params = torch.tensor([[0.1, 0.1]] * noisy.size(0), dtype=torch.float32).to(device)
            denoised = model(noisy, params=params, iterations=10)
            denoised = denoised.cpu().numpy()
            clean = clean.cpu().numpy()
            for i in range(denoised.shape[0]):
                clean_img = clean[i].transpose(1,2,0)         # [H, W, C]
                denoised_img = denoised[i].transpose(1,2,0)  # [H, W, C]
                h, w, c = clean_img.shape
                win_size = 7
                if h < win_size or w < win_size:
                    win_size = min(h, w) if min(h, w) % 2 == 1 else min(h, w) - 1
                p = psnr(clean_img, denoised_img, data_range=1)
                s = ssim(clean_img, denoised_img, data_range=1, channel_axis=2, win_size=win_size)
                a = compute_sam(clean_img, denoised_img)
                psnr_list.append(p)
                ssim_list.append(s)
                sam_list.append(a)
    return np.mean(psnr_list), np.mean(ssim_list), np.mean(sam_list)

def evaluate_with_bm3d_alternative(test_loader, sigma=0.1):
    """
    Evaluate BM3D denoising on the test set using the bm3d package.

    Parameters:
        test_loader (DataLoader): DataLoader for the test set.
        sigma (float): Noise standard deviation.

    Returns:
        mean_psnr, mean_ssim, mean_sam: Evaluation metrics.
    """
    psnr_list, ssim_list, sam_list = [], [], []
    for noisy, clean in test_loader:
        noisy = noisy.numpy()
        clean = clean.numpy()
        denoised = []
        for i in range(noisy.shape[0]):
            try:
                denoised_img = bm3d_denoise_alternative(noisy[i], sigma)
                denoised.append(denoised_img)
            except ValueError as e:
                print(f"Skipping patch {i} due to error: {e}")
                denoised.append(noisy[i])  # Optionally, append the noisy image if denoising fails
        denoised = np.array(denoised)
        for i in range(len(denoised)):
            clean_img = clean[i].transpose(1,2,0)         # [H, W, C]
            denoised_img = denoised[i].transpose(1,2,0)  # [H, W, C]
            h, w, c = clean_img.shape
            win_size = 7
            if h < win_size or w < win_size:
                win_size = min(h, w) if min(h, w) % 2 == 1 else min(h, w) - 1
            p = psnr(clean_img, denoised_img, data_range=1)
            s = ssim(clean_img, denoised_img, data_range=1, channel_axis=2, win_size=win_size)
            a = compute_sam(clean_img, denoised_img)
            psnr_list.append(p)
            ssim_list.append(s)
            sam_list.append(a)
    return np.mean(psnr_list), np.mean(ssim_list), np.mean(sam_list)

def evaluate_models(models, model_names, device, test_loader):
    """
    Evaluate multiple models on the test set and compile their metrics.

    Parameters:
        models (list): List of models to evaluate.
        model_names (list): Corresponding names of the models.
        device (torch.device): Device to run computations on.
        test_loader (DataLoader): DataLoader for the test set.

    Returns:
        metrics_dict (dict): Dictionary containing metrics for each model.
    """
    metrics_dict = {}
    for model, name in zip(models, model_names):
        if name == 'BM3D':
            psnr_val, ssim_val, sam_val = evaluate_with_bm3d_alternative(test_loader, sigma=0.1)
        else:
            psnr_val, ssim_val, sam_val = evaluate_model(model, model.denoiser, device, test_loader)
        metrics_dict[name] = {'PSNR': psnr_val, 'SSIM': ssim_val, 'SAM': sam_val}
        logging.info(f"{name} Evaluation: PSNR: {psnr_val:.2f}, SSIM: {ssim_val:.4f}, SAM: {sam_val:.4f}")
    return metrics_dict

def generate_evaluation_table(metrics_dict):
    """
    Generate a pandas DataFrame from the metrics dictionary.

    Parameters:
        metrics_dict (dict): Dictionary containing metrics for each model.

    Returns:
        df (DataFrame): Pandas DataFrame with evaluation metrics.
    """
    df = pd.DataFrame(metrics_dict).T  # Transpose to have models as rows
    return df

def plot_metrics(df_metrics):
    """
    Plot evaluation metrics using seaborn.

    Parameters:
        df_metrics (DataFrame): DataFrame containing evaluation metrics.
    """
    df_melted = df_metrics.reset_index().melt(id_vars='index', var_name='Metric', value_name='Value')
    df_melted.rename(columns={'index': 'Model'}, inplace=True)

    plt.figure(figsize=(12, 6))
    sns.barplot(x='Metric', y='Value', hue='Model', data=df_melted)
    plt.title('Comparison of Evaluation Metrics Across Models', fontsize=16)
    plt.ylabel('Metric Value', fontsize=14)
    plt.xlabel('Evaluation Metric', fontsize=14)
    plt.legend(title='Model', fontsize=12)
    plt.tight_layout()
    plt.show()

# ----------------------------
# 7a. BM3D Denoising Function (Alternative)
# ----------------------------
# Import bm3d with alias to prevent conflicts with bm4d
from bm3d import bm3d as bm3d_denoise

def bm3d_denoise_alternative(noisy_image, sigma):
    """
    Apply BM3D denoising using the bm3d package.

    Parameters:
        noisy_image (numpy.ndarray): Noisy image with shape (H, W, C).
        sigma (float): Noise standard deviation.

    Returns:
        denoised_image (numpy.ndarray): Denoised image.
    """
    denoised = np.zeros_like(noisy_image)
    for c in range(noisy_image.shape[2]):
        h, w = noisy_image[:, :, c].shape
        print(f"Denoising channel {c}: shape {h}x{w}")
        if h < 8 or w < 8:
            raise ValueError(f"Patch size for channel {c} is too small: {h}x{w}")
        denoised[:, :, c] = bm3d_denoise(noisy_image[:, :, c], sigma)
    return denoised

# ----------------------------
# 11. Training Function
# ----------------------------
def train_models(admm_tfpnp, admm_qrnn3d, ppo_tfpnp, ppo_qrnn3d, optimizer_tfpnp, optimizer_qrnn3d,
                train_loader, test_loader, device, epochs=50):
    """
    Train ADMM Solvers with PPO agents.

    Parameters:
        admm_tfpnp (ADMMSolver): ADMM Solver with TFPNP denoiser.
        admm_qrnn3d (ADMMSolver): ADMM Solver with QRNN3D denoiser.
        ppo_tfpnp (PPO): PPO agent for TFPNP.
        ppo_qrnn3d (PPO): PPO agent for QRNN3D.
        optimizer_tfpnp (torch.optim.Optimizer): Optimizer for TFPNP.
        optimizer_qrnn3d (torch.optim.Optimizer): Optimizer for QRNN3D.
        train_loader (DataLoader): DataLoader for the training set.
        test_loader (DataLoader): DataLoader for the test set.
        device (torch.device): Device to run computations on.
        epochs (int): Number of training epochs.

    Returns:
        None
    """
    # Initialize lists to track PPO agent's parameters
    sigma_d_values = []
    mu_values = []

    # Initialize lists to store metrics
    training_losses = []
    admm_psnr = []
    admm_ssim = []
    admm_sam = []
    bm3d_psnr = []
    bm3d_ssim = []
    bm3d_sam = []

    for epoch in range(epochs):
        logging.info(f"\n=== Epoch {epoch + 1}/{epochs} ===")
        # ------------------------
        # Train TFPNP
        # ------------------------
        admm_tfpnp.train()
        total_loss_tfpnp = 0
        for noisy, clean in train_loader:
            noisy, clean = noisy.to(device), clean.to(device)
            logging.debug(f"TFPNP - Noisy shape: {noisy.shape}, Clean shape: {clean.shape}")  # Debugging

            # Get action from PPO
            state = np.zeros((noisy.size(0), 1), dtype=np.float32)  # Placeholder state
            actions, _ = ppo_tfpnp.predict(state, deterministic=True)
            sigma_d = actions[:, 0]
            mu = actions[:, 1]
            tau = np.clip(actions[:, 2], 1, admm_tfpnp.max_iterations).astype(int)

            # Prepare parameters tensor
            params = torch.tensor(list(zip(sigma_d, mu)), dtype=torch.float32).to(device)
            logging.debug(f"TFPNP - Params shape: {params.shape}")  # Debugging

            # Forward pass through ADMM solver
            denoised = admm_tfpnp(noisy, params=params, iterations=tau.tolist())
            logging.debug(f"TFPNP - Denoised shape: {denoised.shape}")  # Debugging

            # Compute loss
            loss = nn.MSELoss()(denoised, clean)
            loss.backward()

            # Update models
            optimizer_tfpnp.step()
            optimizer_tfpnp.zero_grad()

            total_loss_tfpnp += loss.item()

            # Track PPO agent's parameters
            sigma_d_values.extend(sigma_d.tolist())
            mu_values.extend(mu.tolist())

        avg_loss_tfpnp = total_loss_tfpnp / len(train_loader)
        training_losses.append(avg_loss_tfpnp)
        logging.info(f"TFPNP Training Loss: {avg_loss_tfpnp:.4f}")

        # Update PPO for TFPNP
        ppo_tfpnp.learn(total_timesteps=1000)

        # ------------------------
        # Train QRNN3D
        # ------------------------
        admm_qrnn3d.train()
        total_loss_qrnn3d = 0
        for noisy, clean in train_loader:
            noisy, clean = noisy.to(device), clean.to(device)
            logging.debug(f"QRNN3D - Noisy shape: {noisy.shape}, Clean shape: {clean.shape}")  # Debugging

            # Get action from PPO
            state = np.zeros((noisy.size(0), 1), dtype=np.float32)  # Placeholder state
            actions, _ = ppo_qrnn3d.predict(state, deterministic=True)
            sigma_d = actions[:, 0]
            mu = actions[:, 1]
            tau = np.clip(actions[:, 2], 1, admm_qrnn3d.max_iterations).astype(int)

            # Prepare parameters tensor
            params = torch.tensor(list(zip(sigma_d, mu)), dtype=torch.float32).to(device)
            logging.debug(f"QRNN3D - Params shape: {params.shape}")  # Debugging

            # Forward pass through ADMM solver
            denoised = admm_qrnn3d(noisy, params=params, iterations=tau.tolist())
            logging.debug(f"QRNN3D - Denoised shape: {denoised.shape}")  # Debugging

            # Compute loss
            loss = nn.MSELoss()(denoised, clean)
            loss.backward()

            # Update models
            optimizer_qrnn3d.step()
            optimizer_qrnn3d.zero_grad()

            total_loss_qrnn3d += loss.item()

            # Track PPO agent's parameters
            sigma_d_values.extend(sigma_d.tolist())
            mu_values.extend(mu.tolist())

        avg_loss_qrnn3d = total_loss_qrnn3d / len(train_loader)
        training_losses.append(avg_loss_qrnn3d)
        logging.info(f"QRNN3D Training Loss: {avg_loss_qrnn3d:.4f}")

        # Update PPO for QRNN3D
        ppo_qrnn3d.learn(total_timesteps=1000)

        # ------------------------
        # Evaluation every 10 epochs
        # ------------------------
        if (epoch + 1) % 10 == 0 or epoch == 0:
            logging.info(f"\n--- Evaluation at Epoch {epoch + 1} ---")
            # Evaluate ADMM Solver with TFPNP
            mean_psnr_tfpnp, mean_ssim_tfpnp, mean_sam_tfpnp = evaluate_model(admm_tfpnp, admm_tfpnp.denoiser, device, test_loader)
            admm_psnr.append(mean_psnr_tfpnp)
            admm_ssim.append(mean_ssim_tfpnp)
            admm_sam.append(mean_sam_tfpnp)
            logging.info(f"TFPNP - PSNR: {mean_psnr_tfpnp:.2f}, SSIM: {mean_ssim_tfpnp:.4f}, SAM: {mean_sam_tfpnp:.4f}")

            # Evaluate ADMM Solver with QRNN3D
            mean_psnr_qrnn3d, mean_ssim_qrnn3d, mean_sam_qrnn3d = evaluate_model(admm_qrnn3d, admm_qrnn3d.denoiser, device, test_loader)
            admm_psnr.append(mean_psnr_qrnn3d)
            admm_ssim.append(mean_ssim_qrnn3d)
            admm_sam.append(mean_sam_qrnn3d)
            logging.info(f"QRNN3D - PSNR: {mean_psnr_qrnn3d:.2f}, SSIM: {mean_ssim_qrnn3d:.4f}, SAM: {mean_sam_qrnn3d:.4f}")

            # Evaluate BM3D as a benchmark
            mean_psnr_bm3d, mean_ssim_bm3d, mean_sam_bm3d = evaluate_with_bm3d_alternative(test_loader, sigma=0.1)
            bm3d_psnr.append(mean_psnr_bm3d)
            bm3d_ssim.append(mean_ssim_bm3d)
            bm3d_sam.append(mean_sam_bm3d)
            logging.info(f"BM3D - PSNR: {mean_psnr_bm3d:.2f}, SSIM: {mean_ssim_bm3d:.4f}, SAM: {mean_sam_bm3d:.4f}")

            # Compile metrics
            metrics = {
                'TFPNP': {'PSNR': mean_psnr_tfpnp, 'SSIM': mean_ssim_tfpnp, 'SAM': mean_sam_tfpnp},
                'QRNN3D': {'PSNR': mean_psnr_qrnn3d, 'SSIM': mean_ssim_qrnn3d, 'SAM': mean_sam_qrnn3d},
                'BM3D': {'PSNR': mean_psnr_bm3d, 'SSIM': mean_ssim_bm3d, 'SAM': mean_sam_bm3d}
            }
            df_metrics = generate_evaluation_table(metrics)
            logging.info("\nEvaluation Metrics:")
            print(df_metrics)

            # Plot Metrics
            plot_metrics(df_metrics)

            # Optional: Visualize sample denoising results at evaluation epochs
            with torch.no_grad():
                # Fetch a small batch for visualization
                vis_noisy, vis_clean = next(iter(test_loader))
                vis_noisy = vis_noisy.to(device)
                vis_clean = vis_clean.numpy()

                # Get actions for visualization batch (TFPNP)
                vis_observations = np.zeros((vis_noisy.size(0), 1), dtype=np.float32)  # Placeholder state
                vis_actions_tfpnp, _ = ppo_tfpnp.predict(vis_observations, deterministic=True)
                vis_actions_tfpnp[:, 2] = np.clip(vis_actions_tfpnp[:, 2], 1, admm_tfpnp.max_iterations)
                vis_actions_tfpnp[:, 0] = np.clip(vis_actions_tfpnp[:, 0], 0.0, 1.0)
                vis_actions_tfpnp[:, 1] = np.clip(vis_actions_tfpnp[:, 1], 0.0, 1.0)

                # Extract actions
                vis_sigma_d_tfpnp = vis_actions_tfpnp[:, 0].astype(float)
                vis_mu_tfpnp = vis_actions_tfpnp[:, 1].astype(float)
                vis_tau_tfpnp = np.clip(vis_actions_tfpnp[:, 2], 1, admm_tfpnp.max_iterations).astype(int)

                # Prepare parameters tensor
                vis_params_tfpnp = torch.tensor(list(zip(vis_sigma_d_tfpnp, vis_mu_tfpnp)), dtype=torch.float32).to(device)

                # Forward pass through ADMM solver (TFPNP)
                denoised_vis_tfpnp = admm_tfpnp(vis_noisy, params=vis_params_tfpnp, iterations=vis_tau_tfpnp.tolist()).cpu().numpy()

                # Get actions for QRNN3D
                vis_actions_qrnn3d, _ = ppo_qrnn3d.predict(vis_observations, deterministic=True)
                vis_actions_qrnn3d[:, 2] = np.clip(vis_actions_qrnn3d[:, 2], 1, admm_qrnn3d.max_iterations)
                vis_actions_qrnn3d[:, 0] = np.clip(vis_actions_qrnn3d[:, 0], 0.0, 1.0)
                vis_actions_qrnn3d[:, 1] = np.clip(vis_actions_qrnn3d[:, 1], 0.0, 1.0)

                # Extract actions
                vis_sigma_d_qrnn3d = vis_actions_qrnn3d[:, 0].astype(float)
                vis_mu_qrnn3d = vis_actions_qrnn3d[:, 1].astype(float)
                vis_tau_qrnn3d = np.clip(vis_actions_qrnn3d[:, 2], 1, admm_qrnn3d.max_iterations).astype(int)

                # Prepare parameters tensor
                vis_params_qrnn3d = torch.tensor(list(zip(vis_sigma_d_qrnn3d, vis_mu_qrnn3d)), dtype=torch.float32).to(device)

                # Forward pass through ADMM solver (QRNN3D)
                denoised_vis_qrnn3d = admm_qrnn3d(vis_noisy, params=vis_params_qrnn3d, iterations=vis_tau_qrnn3d.tolist()).cpu().numpy()

                # Visualize the first sample in the batch
                visualize_results(
                    noisy=noisy.cpu().numpy(),
                    clean=vis_clean,
                    denoised_tfpnp=denoised_vis_tfpnp,
                    denoised_qrnn3d=denoised_vis_qrnn3d,
                    sample=0,
                    bands=[0,1,2]
                )

def plot_tracked_parameters(sigma_d_values, mu_values):
    """
    Plot tracked parameters over training.

    Parameters:
        sigma_d_values (list): Tracked sigma_d values.
        mu_values (list): Tracked mu values.
    """
    plt.figure(figsize=(10, 5))
    plt.plot(sigma_d_values, label='sigma_d', color='green')
    plt.plot(mu_values, label='mu', color='red')
    plt.title('Tracked Parameters Over Training', fontsize=16)
    plt.xlabel('Iteration', fontsize=14)
    plt.ylabel('Parameter Value', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    plt.savefig('tracked_parameters_over_training.png', dpi=300)

# ----------------------------
# 13. Final Evaluation and Visualization
# ----------------------------
def final_evaluation_and_visualization(admm_tfpnp, admm_qrnn3d, test_loader, device, ppo_tfpnp, ppo_qrnn3d, sigma_d_values, mu_values):
    """
    Perform final evaluation and visualize denoising results.

    Parameters:
        admm_tfpnp (ADMMSolver): ADMM Solver with TFPNP denoiser.
        admm_qrnn3d (ADMMSolver): ADMM Solver with QRNN3D denoiser.
        test_loader (DataLoader): DataLoader for the test set.
        device (torch.device): Device to run computations on.
        ppo_tfpnp (PPO): PPO agent for TFPNP.
        ppo_qrnn3d (PPO): PPO agent for QRNN3D.
        sigma_d_values (list): Tracked sigma_d values.
        mu_values (list): Tracked mu values.
    """
    logging.info("\n--- Final Evaluation ---")
    # Evaluate TFPNP
    psnr_tfpnp_val, ssim_tfpnp_val, sam_tfpnp_val = evaluate_model(admm_tfpnp, admm_tfpnp.denoiser, device, test_loader)
    logging.info(f"TFPNP - PSNR: {psnr_tfpnp_val:.2f}, SSIM: {ssim_tfpnp_val:.4f}, SAM: {sam_tfpnp_val:.4f}")

    # Evaluate QRNN3D
    psnr_qrnn3d_val, ssim_qrnn3d_val, sam_qrnn3d_val = evaluate_model(admm_qrnn3d, admm_qrnn3d.denoiser, device, test_loader)
    logging.info(f"QRNN3D - PSNR: {psnr_qrnn3d_val:.2f}, SSIM: {ssim_qrnn3d_val:.4f}, SAM: {sam_qrnn3d_val:.4f}")

    # Evaluate BM3D
    mean_psnr_bm3d, mean_ssim_bm3d, mean_sam_bm3d = evaluate_with_bm3d_alternative(test_loader, sigma=0.1)
    logging.info(f"BM3D - PSNR: {mean_psnr_bm3d:.2f}, SSIM: {mean_ssim_bm3d:.4f}, SAM: {mean_sam_bm3d:.4f}")

    # Compile metrics
    metrics = {
        'TFPNP': {'PSNR': psnr_tfpnp_val, 'SSIM': ssim_tfpnp_val, 'SAM': sam_tfpnp_val},
        'QRNN3D': {'PSNR': psnr_qrnn3d_val, 'SSIM': ssim_qrnn3d_val, 'SAM': sam_qrnn3d_val},
        'BM3D': {'PSNR': mean_psnr_bm3d, 'SSIM': mean_ssim_bm3d, 'SAM': mean_sam_bm3d}
    }
    df_metrics = generate_evaluation_table(metrics)
    logging.info("\nEvaluation Metrics:")
    print(df_metrics)

    # Plot Metrics
    plot_metrics(df_metrics)

    # ----------------------------
    # Final Visualization
    # ----------------------------
    # Get a batch from test_loader
    noisy, clean = next(iter(test_loader))
    noisy = noisy.to(device)
    clean = clean.numpy()

    # Get actions from PPO agent for the batch (TFPNP)
    observations = np.zeros((noisy.size(0), 1), dtype=np.float32)  # Placeholder state
    actions_tfpnp, _ = ppo_tfpnp.predict(observations, deterministic=True)
    actions_tfpnp[:, 2] = np.clip(actions_tfpnp[:, 2], 1, admm_tfpnp.max_iterations)
    actions_tfpnp[:, 0] = np.clip(actions_tfpnp[:, 0], 0.0, 1.0)
    actions_tfpnp[:, 1] = np.clip(actions_tfpnp[:, 1], 0.0, 1.0)

    # Extract actions
    sigma_d_tfpnp = actions_tfpnp[:, 0].astype(float)
    mu_tfpnp = actions_tfpnp[:, 1].astype(float)
    tau_tfpnp = np.clip(actions_tfpnp[:, 2], 1, admm_tfpnp.max_iterations).astype(int)

    # Prepare parameters tensor
    params_tfpnp = torch.tensor(list(zip(sigma_d_tfpnp, mu_tfpnp)), dtype=torch.float32).to(device)

    # Forward pass through ADMM solver (TFPNP)
    with torch.no_grad():
        denoised_tfpnp = admm_tfpnp(noisy, params=params_tfpnp, iterations=tau_tfpnp.tolist()).cpu().numpy()

    # Get actions from PPO agent for the batch (QRNN3D)
    actions_qrnn3d, _ = ppo_qrnn3d.predict(observations, deterministic=True)
    actions_qrnn3d[:, 2] = np.clip(actions_qrnn3d[:, 2], 1, admm_qrnn3d.max_iterations)
    actions_qrnn3d[:, 0] = np.clip(actions_qrnn3d[:, 0], 0.0, 1.0)
    actions_qrnn3d[:, 1] = np.clip(actions_qrnn3d[:, 1], 0.0, 1.0)

    # Extract actions
    sigma_d_qrnn3d = actions_qrnn3d[:, 0].astype(float)
    mu_qrnn3d = actions_qrnn3d[:, 1].astype(float)
    tau_qrnn3d = np.clip(actions_qrnn3d[:, 2], 1, admm_qrnn3d.max_iterations).astype(int)

    # Prepare parameters tensor
    params_qrnn3d = torch.tensor(list(zip(sigma_d_qrnn3d, mu_qrnn3d)), dtype=torch.float32).to(device)

    # Forward pass through ADMM solver (QRNN3D)
    with torch.no_grad():
        denoised_qrnn3d = admm_qrnn3d(noisy, params=params_qrnn3d, iterations=tau_qrnn3d.tolist()).cpu().numpy()

    # Visualize the first sample in the batch
    visualize_results(
        noisy=noisy.cpu().numpy(),
        clean=clean,
        denoised_tfpnp=denoised_tfpnp,
        denoised_qrnn3d=denoised_qrnn3d,
        sample=0,
        bands=[0,1,2]
    )

    # Visualization of multiple samples
    visualize_batch_samples(noisy.cpu().numpy(), clean, denoised_tfpnp, bands=[0,1,2], num_samples=3)

    # ------------------------
    # Plot Tracked Parameters
    # ------------------------
    plot_tracked_parameters(sigma_d_values, mu_values)

    # ------------------------
    # Save a Sample Visualization
    # ------------------------
    save_visualization(noisy.cpu().numpy(), clean, denoised_tfpnp, bands=[0,1,2], sample_idx=0, filename='denoising_result_sample1.png')

    # ------------------------
    # Save Models
    # ------------------------
    torch.save(admm_tfpnp.state_dict(), 'admm_tfpnp.pth')
    torch.save(admm_qrnn3d.state_dict(), 'admm_qrnn3d.pth')
    ppo_tfpnp.save("ppo_tfpnp")
    ppo_qrnn3d.save("ppo_qrnn3d")
    logging.info("Models saved successfully.")

# ----------------------------
# Execute the Entire Process
# ----------------------------

def main():
    # ------------------------
    # 14.1. Hyperparameters
    # ------------------------
    epochs = 50
    learning_rate = 1e-4
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    logging.info(f"Using device: {device}")

    # ------------------------
    # 14.2. Dataset Preparation
    # ------------------------
    # Load Indian Pines with Gaussian noise
    noisy_patches_ip, clean_patches_ip = download_and_preprocess_data(
        noise_type='gaussian',
        noise_params={'noise_std': 0.1}
    )

    # Combine datasets (if multiple)
    # For simplicity, we'll proceed with Indian Pines only
    combined_noisy = noisy_patches_ip
    combined_clean = clean_patches_ip

    # Split into train and test sets (80% train, 20% test)
    train_noisy, test_noisy, train_clean, test_clean = train_test_split(
        combined_noisy, combined_clean, test_size=0.2, random_state=42
    )

    # Create Datasets and DataLoaders with data augmentation for training
    train_dataset = HSIDataset(train_noisy, train_clean, augment=True)
    test_dataset = HSIDataset(test_noisy, test_clean, augment=False)
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
    logging.info(f"Train dataset size: {len(train_dataset)}, Test dataset size: {len(test_dataset)}")

    # ------------------------
    # 14.3. Initialize Models and PPO Agents
    # ------------------------
    admm_tfpnp, admm_qrnn3d, ppo_tfpnp, ppo_qrnn3d, optimizer_tfpnp, optimizer_qrnn3d = initialize_models_and_agents(device, test_loader)

    # ------------------------
    # 14.4. Training Loop
    # ------------------------
    train_models(admm_tfpnp, admm_qrnn3d, ppo_tfpnp, ppo_qrnn3d, optimizer_tfpnp, optimizer_qrnn3d,
                train_loader, test_loader, device, epochs=epochs)

    # ------------------------
    # 16. Final Evaluation and Visualization
    # ------------------------
    # Not needed here as it's already handled within the training loop every 10 epochs.

if __name__ == "__main__":
    main()
```
